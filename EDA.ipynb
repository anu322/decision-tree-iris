```python
df1=df.groupby("Sub-Category")['Profit'].sum().reset_index().sort_values(ascending=False,by='Profit')
df1.reset_index(inplace=True,drop=True)
df2=df1.head(5)
```

```python
plt.figure(figsize=(12,8))
sns.catplot('Sub-Category','Profit',data=df2,kind="bar",height=8)
plt.title("top 5 sub-categories having high profit")
```

```python
df_city=df.groupby("City")['Profit'].sum().reset_index().sort_values(ascending=False,by='Profit')
df_city.reset_index(inplace=True,drop=True)
df_city_=df_city.head(5)
```

```python
plt.figure(figsize=(12,8))
sns.catplot('City','Profit',data=df_city_,kind="bar",height=8)
plt.title("top 5 cities having high profit")
```

```python
df_state=df.groupby("State")['Profit'].sum().reset_index().sort_values(ascending=False,by='Profit')
df_state.reset_index(inplace=True,drop=True)
df_state_=df_state.head(5)
plt.figure(figsize=(12,8))
sns.catplot('State','Profit',data=df_state_,kind="bar",height=8)
plt.title("top 5 states having high profit")
```

```python
df_sales=df.groupby("Sub-Category")['Sales'].sum().reset_index().sort_values(ascending=False,by='Sales')
df_sales.reset_index(inplace=True,drop=True)
df_sales_=df_sales.head(5)
```

```python
plt.figure(figsize=(12,8))
sns.catplot('Sub-Category','Sales',data=df_sales_,kind="bar",height=8)
plt.title("top 5 cities having high sales")
```
